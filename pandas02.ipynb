```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('auto_new.csv')
#print(df)

hp_st_corr,p_value = stats.pearsonr(df["horsepower"],df["stroke"])

print("NOTE\np_value < 0.001")
print("   선형 상관관계 강한 증거")
print("p_value < 0.05")
print("   선형 상관관계 보통 증거")
print("p_value < 0.1")
print("   선형 상관관계 약한 증거\n")
print("horse power과 stroke의 상관계수 :", hp_st_corr)
print("p_value :", p_value,"\n")

Y = df["price"]
Z = df[["horsepower","stroke"]]

lm=LinearRegression()

lm.fit(Z,Y)
Yhat = lm.predict(Z)

print("LinearRegression.intercept_ :",lm.intercept_)
print("LinearRegression.coef_ :",lm.coef_)

print("\n\nRMSE\nmean_squared_error(Y,Yhat):",mean_squared_error(Y,Yhat),"\n해당 수치가 낮을수록 정확도가 높다.\n\n")

sns.regplot(x = "stroke",y = "price",data=df)
plt.ylim(0,)
plt.title("Regression Plot: Price and Stroke")
plt.show()
```
